```python

	

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ১. ডেটাসেট লোড করা (আপনার স্ক্রিনশট অনুযায়ী পাথ)
df = pd.read_csv('/kaggle/input/datasets/litonislam/hr-analytics-cleaned-and-processed-employee-dataset/HR_Data_Cleaned.csv')

# ফিগার সাইজ সেট করা
plt.figure(figsize=(15, 10))

# ২. ওভারটাইম বনাম অ্যাট্রিশন (Bar Plot)
plt.subplot(2, 2, 1)
sns.countplot(data=df, x='OverTime', hue='Attrition', palette=['#228B22', '#DC143C'])
plt.title('Impact of Overtime on Attrition', fontweight='bold')

# ৩. ওভারঅল রিটেনশন রেট (Donut Chart)
plt.subplot(2, 2, 2)
counts = df['Attrition'].value_counts()
plt.pie(counts, autopct='%1.1f%%', startangle=90, colors=['#228B22', '#DC143C'], wedgeprops={'width':0.4})
plt.title('Overall Retention Rate', fontweight='bold')

# ৪. জেন্ডার ডিস্ট্রিবিউশন (Pie Chart)
plt.subplot(2, 2, 3)
gender = df['Gender'].value_counts()
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', colors=['#4169E1', '#87CEEB'])
plt.title('Gender Distribution', fontweight='bold')

# ৫. এমপ্লয়ি বাই ডিপার্টমেন্ট (Horizontal Count Plot)
# এখানে hue='Department' এবং legend=False যোগ করেছি যাতে FutureWarning না আসে
plt.subplot(2, 2, 4)
sns.countplot(data=df, y='Department', hue='Department', palette='viridis', legend=False)
plt.title('Employees by Department', fontweight='bold')

# গ্রাফগুলো সুন্দরভাবে সাজানোর জন্য
plt.tight_layout()

# ৬. আউটপুটে ছবি হিসেবে সেভ করা (যাতে Output ট্যাবে ফাইলটি পান)
plt.savefig('hr_analytics_dashboard.png')

# ৭. সরাসরি নোটবুকে গ্রাফ প্রদর্শন
plt.show()
```
